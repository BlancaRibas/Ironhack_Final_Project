```python
import streamlit as st
# To make things easier later, we're also importing numpy and pandas for
# working with sample data.
import numpy as np
import pandas as pd
from PIL import Image
from google_currency import convert
import pandas as pd
import requests
import cv2
from streamlit_functions import *
import time
import json
import streamlit as st
import os


imagen = Image.open("bann.JPG")
st.image(imagen)
st.markdown('**Convert currencies from pictures!**:sunglasses:')


st.sidebar.markdown('**:one:Input options**:rocket:')
currency_list = ['AUD', 'BGN', 'BRL', 'CAD', 'CHF', 'CNY', 'CZK', 'DKK', 'EUR', 'GBP', 'HKD', 'HRK', 'HUF', 'IDR', 'ILS',
                 'INR', 'ISK', 'JPY', 'KRW', 'MXN', 'MYR', 'NOK', 'NZD', 'PHP', 'PLN', 'RON', 'RUB', 'SEK', 'SGD',
                 'THB', 'TRY', 'USD', 'ZAR']

input_price = st.sidebar.selectbox('Select base currency', currency_list)
if len(input_price) == 0 or input_price == "Select base currency":
    st.write("Please select currency")


output_price = st.sidebar.selectbox('Select target currency', currency_list)
if len(output_price) == 0 or output_price == "Select target currency":
    st.write("Please select currency")

st.sidebar.markdown(':two: :camera:')
uploaded_file = st.sidebar.file_uploader("Upload image", type = ['jpeg', 'jpg', 'png'])

if uploaded_file:
    #st.write('File successfully uploaded')
    img = Image.open(uploaded_file)
    img.save('imagenes/a.jpg')
    st.write('**This is your image!**:zap:')
    st.image(img)
    #if st.button('converting'):

    # os.system('open analisis')
    # cv2.imwrite('analisis1',img)
    image1 = cv2.imread('imagenes/a.jpg')
    # transformacion a escala de grises
    gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    # Creamos copia de la imagen de entrada
    image_aux = gray.copy()
    try:
        x=pytesseract.image_to_string(image_aux).split("\n")[0].replace(",",".")
        y = float(x)
        status_text = st.empty()
        for i in range(100):
            time.sleep(0.0001)
        status_text.text('thinking')
        st.balloons()

        yas = converter(input_price, output_price, y)
        st.write('**Converted!**:money_with_wings:')
        st.write(yas)
    except:
         st.write("sorry, couldn't read image, try again!:ghost:")
    #st.write(x)

#---------------------------------#
# About
expander_bar = st.beta_expander("About")
expander_bar.markdown("""
* **Python libraries:** streamlit, pandas, pillow, numpy, requests, json, pytesseract, google_currency, OpenCV
* **Credit:** App built by Blanca Ribas Pombo
* **Github:** https://github.com/BlancaRibas
* ** Ironhack Final Project, June 2021
""")
```
