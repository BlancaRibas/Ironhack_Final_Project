```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt        
```

```python
plt.imshow(img_cv)
```

```python
img_rgb= cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
image_bw = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY )# convert image to black and white
```

```python
plt.imshow(image_bw)
```
